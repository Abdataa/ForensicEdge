```python
from pathlib import Path

base = Path("/content/drive/MyDrive/ForensicEdge/ai_engine/datasets/raw/fingerprints/SOCOFing")

for path in base.rglob("Real"):
    print(f"\nChecking: {path}")
    items = list(path.iterdir())
    print(f"Total items: {len(items)}")

    # Show first few items
    for item in items[:5]:
        print("  ", item.name, "(DIR)" if item.is_dir() else "(FILE)")
```

```python
from pathlib import Path
import shutil

outer = Path("/content/drive/MyDrive/ForensicEdge/ai_engine/datasets/raw/fingerprints/SOCOFing/Real")
inner = outer / "Real"

# Move all files from inner → outer
for file in inner.iterdir():
    if file.is_file():
        shutil.move(str(file), str(outer / file.name))

print("Files moved successfully.")

# Remove the now-empty inner folder
inner.rmdir()

print("Empty inner 'Real' folder removed.")
```

```python
from pathlib import Path

BASE = Path("/content/drive/MyDrive/ForensicEdge/ai_engine/datasets/processed/SOCOFing")

identities = [d for d in BASE.iterdir() if d.is_dir()]

print("Total identity folders:", len(identities))

all_counts = []

for identity in identities:
    num_files = len([f for f in identity.iterdir() if f.is_file()])
    all_counts.append(num_files)

print("\nMin images per identity:", min(all_counts))
print("Max images per identity:", max(all_counts))
print("Average images per identity:", sum(all_counts)/len(all_counts))
```
